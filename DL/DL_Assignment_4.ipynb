**Question 1:**

**How would you describe TensorFlow in a short sentence? What are its main features? Can you name other popular Deep Learning libraries?**

TensorFlow is an open-source machine learning framework that facilitates the development and deployment of various machine learning and deep learning models. Its main features include a flexible computational graph, automatic differentiation, support for GPUs and TPUs, built-in libraries for neural networks, and tools for model deployment. Other popular deep learning libraries include PyTorch, Keras (which is now integrated with TensorFlow), and MXNet.

**Question 2:**

**Is TensorFlow a drop-in replacement for NumPy? What are the main differences between the two?**

TensorFlow is not a drop-in replacement for NumPy, although it shares some similarities. While both libraries deal with multi-dimensional arrays, TensorFlow introduces the concept of computational graphs and supports automatic differentiation for gradient-based optimization in machine learning. TensorFlow's tensors are designed to be used in graph-based computations for efficient execution on GPUs and TPUs, making it more suitable for deep learning tasks.

**Question 4:**

**Can you name six other data structures available in TensorFlow, beyond regular tensors?**

Beyond regular tensors, TensorFlow offers various specialized data structures, including SparseTensors, RaggedTensors, TensorArray, Dataset, FeatureColumn, and Queue.

**Question 5:**

**A custom loss function can be defined by writing a function or by subclassing the keras.losses.Loss class. When would you use each option?**

Defining a custom loss function using a function is suitable for simpler loss computations that don't require additional internal states. Subclassing `keras.losses.Loss` is preferred for more complex loss functions that need to maintain state across batches or epochs.

**Question 6:**

**Similarly, a custom metric can be defined in a function or a subclass of keras.metrics.Metric. When would you use each option?**

Defining a custom metric as a function is useful when the metric can be directly calculated based on predictions and true labels without needing internal states. Subclassing `keras.metrics.Metric` is necessary when the metric calculation involves maintaining state or aggregating values across batches or epochs.

**Question 7:**

**When should you create a custom layer versus a custom model?**

Create a custom layer when you need to define a new type of operation within a model (e.g., custom activation function or custom transformation). Create a custom model when you need to define a completely new architecture or network structure that cannot be achieved by composing existing layers.

**Question 8:**

**What are some use cases that require writing your own custom training loop?**

Writing your own custom training loop is necessary when you require fine-grained control over training dynamics, specialized learning rate schedules, complex loss combinations, or unique optimization techniques that are not supported by high-level APIs like Keras.

**Question 9:**

**Can custom Keras components contain arbitrary Python code, or must they be convertible to TF Functions?**

Custom Keras components must be convertible to TensorFlow's computation graph (TF Functions) for optimal performance. While some Python code can be used within custom components, it's important to follow TensorFlow's guidelines.

**Question 10:**

**What are the main rules to respect if you want a function to be convertible to a TF Function?**

To ensure that a function is convertible to a TF Function, avoid using any Python control flow statements (e.g., if, while), use TensorFlow operations for computations, and avoid using non-TensorFlow libraries or functions within the function.

**Question 11:**

**When would you need to create a dynamic Keras model? How do you do that? Why not make all your models dynamic?**

You would need to create a dynamic Keras model when the architecture of the model changes based on input data or conditions. This can be achieved by using the functional API to create models with variable branch lengths or operations. However, not all models need to be dynamic; static models are simpler to manage and optimize when the architecture remains fixed.
