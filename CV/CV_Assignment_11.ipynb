
**Question 1: What do REGION PROPOSALS entail?**

Region proposals are bounding box predictions that indicate potential object locations in an image. In object detection tasks, region proposal algorithms suggest regions that are likely to contain objects, reducing the search space for the subsequent object classification and localization tasks.

**Question 2: What do you mean by NON-MAXIMUM SUPPRESSION? (NMS)**

Non-Maximum Suppression (NMS) is a post-processing technique used in object detection to eliminate duplicate bounding box predictions. It filters out bounding boxes that have high overlap (measured using IoU) with each other, keeping only the box with the highest confidence score.

**Question 3: What exactly is mAP?**

mAP stands for mean Average Precision. It is a widely used metric for evaluating object detection models. It measures the accuracy of object localization and classification by calculating the average precision (AP) across different levels of confidence thresholds. The mean AP (mAP) provides an overall assessment of the model's performance.

**Question 4: What is a frames per second (FPS)?**

Frames per second (FPS) is a measure of how many individual images (frames) a video or image processing system can process or display in one second. In object detection, FPS indicates how quickly the model can process an image or a video stream.

**Question 5: What is an IOU (INTERSECTION OVER UNION)?**

Intersection over Union (IoU) is a metric used to measure the overlap between two bounding boxes or regions. It is calculated as the area of intersection divided by the area of union between the boxes. IoU is often used to determine how well a predicted bounding box aligns with a ground truth bounding box.

**Question 6: Describe the PRECISION-RECALL CURVE (PR CURVE)**

The Precision-Recall (PR) curve is a graphical representation of the trade-off between precision and recall at different confidence score thresholds. Precision is the fraction of true positive predictions among all positive predictions, and recall is the fraction of true positive predictions among all actual positives. The PR curve shows how changing the threshold affects these metrics.

**Question 7: What is the term "selective search"?**

Selective Search is a region proposal algorithm used in object detection. It generates potential object regions by grouping pixels based on color, texture, and other features. These regions are then proposed as potential bounding box locations for further analysis.

**Question 8: Describe the R-CNN model's four components.**

The R-CNN (Region Convolutional Neural Network) model consists of four main components:
1. **Region Proposals**: Potential object locations generated by selective search or other methods.
2. **Feature Extraction**: Extracts CNN features from each region proposal.
3. **Localization Module**: Predicts the bounding box coordinates for each object.
4. **Classification Module**: Classifies the content of each region into predefined classes.

**Question 9: What exactly is the Localization Module?**

The Localization Module in R-CNN predicts the precise coordinates (x, y, width, height) of the bounding box around an object within a region proposal. It uses the features extracted from the region proposal to make these predictions.

**Question 10: What are the R-CNN DISADVANTAGES?**

R-CNN has several disadvantages:
- Slow: Processing each region proposal independently leads to slow training and inference.
- Computationally Expensive: Extracting CNN features for each region is time-consuming.
- Fixed Output Size: R-CNN requires fixed-size input, which can result in information loss.
- Separate Training: Different components need to be trained independently, leading to inefficiency.
