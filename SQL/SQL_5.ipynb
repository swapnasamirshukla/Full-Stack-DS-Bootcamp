### Q1: List All User Ids who have all the required skills. Required Skills are DS,Python,SQL and Tableau"


```python
import mysql.connector

# Database connection configuration
config = {
    'host': 'localhost',
    'user': 'root',
    'password': 'Nus@2020',
    'database': 'ineuron'
}

# The SQL query
query = """
select * from ats;
"""

try:
    # Create a connection
    connection = mysql.connector.connect(**config)
    
    # Create a cursor
    cursor = connection.cursor()

    # Execute the query
    cursor.execute(query)
    
    # Fetch all results
    results = cursor.fetchall()

    # Print the results
    for row in results:
        print(row)

except mysql.connector.Error as err:
    print(f"Error: {err}")

finally:
    # Close cursor and connection
    cursor.close()
    connection.close()
```

```python
# The SQL query
query = """
with skillsets 
as
(
SELECT 
    id,
    GROUP_CONCAT(tech order by tech asc SEPARATOR ', ') as skills
FROM 
    ats
GROUP BY 
    id)

select id from
(
select id, sum(
case 
when lower(skills) like '%ds, python, sql, tableau%' then 1
else 0
end) as match_skills
from skillsets
group by id)x
where match_skills=1;
"""

try:
    # Create a connection
    connection = mysql.connector.connect(**config)
    
    # Create a cursor
    cursor = connection.cursor()

    # Execute the query
    cursor.execute(query)
    
    # Fetch all results
    results = cursor.fetchall()

    # Print the results
    for row in results:
        print(row)

except mysql.connector.Error as err:
    print(f"Error: {err}")

finally:
    # Close cursor and connection
    cursor.close()
    connection.close()
```
