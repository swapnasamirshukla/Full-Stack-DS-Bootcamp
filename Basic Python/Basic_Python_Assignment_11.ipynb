Question 8:

Answer:
- Logging messages can be controlled based on severity levels.
- Logging messages can be directed to different outputs (console, files, etc.).
- Logging messages can be easily disabled or enabled in a centralized manner.
- Logging messages provide more structured information for debugging and analysis.

Question 9:
What are the differences between the Step Over, Step In, and Step Out buttons in the debugger?

Answer:
- Step Over: Executes the current line and stops at the next line in the same function, without entering into function calls.
- Step In: Enters the function being called on the current line, allowing you to step through its execution.
- Step Out: Continues execution until the current function returns and then stops on the calling line.

Question 10:
After you click Continue, when will the debugger stop?

Answer:
After clicking Continue, the debugger will stop when it encounters the next breakpoint or when the program execution is complete.

Question 11:
What is the concept of a breakpoint?

Answer:
A breakpoint is a specific point in your code where you instruct the debugger to pause execution. It allows you to inspect the program's state, variables, and execution flow at that particular point. Breakpoints are immensely helpful for debugging, as they help you analyze your code's behavior step by step and identify issues.
