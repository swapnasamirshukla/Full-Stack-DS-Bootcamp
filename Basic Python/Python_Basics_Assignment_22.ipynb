### 1. What is the result of the code, and explain?

```python
>>> X = 'iNeuron'
>>> def func():
        print(X)

>>> func()
```

Result: `iNeuron`

Explanation: The function `func()` prints the value of `X`, which is a global variable. The global variable `X` has the value `'iNeuron'`.

### 2. What is the result of the code, and explain?

```python
>>> X = 'iNeuron'
>>> def func():
        X = 'NI!'

>>> func()
>>> print(X)
```

Result: `iNeuron`

Explanation: Even though the function `func()` defines a local variable `X` with the value `'NI!'`, the global variable `X` remains unchanged. The function doesn't return anything or modify the global variable.

### 3. What does this code print, and why?

```python
>>> X = 'iNeuron'
>>> def func():
        X = 'NI'
        print(X)

>>> func()
>>> print(X)
```

Result: `NI`
         `iNeuron`

Explanation: The function `func()` defines a local variable `X` with the value `'NI'` and prints it. After that, the global variable `X` is printed, which remains `'iNeuron'`.

### 4. What output does this code produce? Why?

```python
>>> X = 'iNeuron'
>>> def func():
        global X
        X = 'NI'

>>> func()
>>> print(X)
```

Result: `NI`

Explanation: The `global` keyword tells Python that `X` inside the function refers to the global variable `X`. Thus, the function modifies the global variable to `'NI'`.

### 5. What about this code—what’s the output, and why?

```python
>>> X = 'iNeuron'
>>> def func():
        X = 'NI'
        def nested():
            print(X)
        nested()

>>> func()
>>> X
```

Result: `NI`
         `'iNeuron'`

Explanation: The `nested()` function prints the local `X` from its enclosing function, which is `'NI'`. After that, the global `X` is evaluated, which is `'iNeuron'`.

### 6. How about this code: what is its output in Python 3, and explain?

```python
>>> def func():
        X = 'NI'
        def nested():
            nonlocal X
            X = 'Spam'
        nested()
        print(X)

>>> func()
```

Result: `Spam`

Explanation: The `nonlocal` keyword indicates that `X` refers to the nearest enclosing scope that is not global. It modifies `X` in the nearest enclosing scope, which is the `func()` function. Therefore, `func()` prints `'Spam'`.
