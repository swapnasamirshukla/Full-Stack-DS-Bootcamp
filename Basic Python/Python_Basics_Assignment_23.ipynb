### 1. What is the result of the code, and why?

```python
>>> def func(a, b=6, c=8):
        print(a, b, c)
>>> func(1, 2)
```

Result: `1 2 8`

Explanation: You provided values for `a` and `b`, which are `1` and `2` respectively. The value for `c` is taken from the default argument, which is `8`.

### 2. What is the result of this code, and why?

```python
>>> def func(a, b, c=5):
        print(a, b, c)
>>> func(1, c=3, b=2)
```

Result: `1 2 3`

Explanation: You provided values for all parameters `a`, `b`, and `c` which are `1`, `2`, and `3` respectively.

---

### 3. How about this code: what is its result, and why?

```python
>>> def func(a, *pargs):
        print(a, pargs)
>>> func(1, 2, 3)
```

Result: `1 (2, 3)`

Explanation: `*pargs` collects additional positional arguments into a tuple. The value for `a` is `1`, and the rest `(2, 3)` are packed into `pargs`.

---

### 4. What does this code print, and why?

```python
>>> def func(a, **kargs):
        print(a, kargs)
>>> func(a=1, c=3, b=2)
```

Result: `1 {'c': 3, 'b': 2}`

Explanation: `**kargs` collects additional keyword arguments into a dictionary. The value for `a` is `1`, and the rest `{'c': 3, 'b': 2}` are packed into `kargs`.

---

### 5. What gets printed by this, and explain?

```python
>>> def func(a, b, c=8, d=5): print(a, b, c, d)
>>> func(1, *(5, 6))
```

Result: `1 5 6 5`

Explanation: `*(5, 6)` unpacks the tuple and provides values for `b` and `c` which are `5` and `6`. The values for `a` and `d` are `1` and the default `5`, respectively.

### 6. what is the result of this, and explain?

```python
>>> def func(a, b, c): a = 2; b[0] = 'x'; c['a'] = 'y'
>>> l=1; m=[1]; n={'a':0}
>>> func(l, m, n)
>>> l, m, n
```

Result: `(1, ['x'], {'a': 'y'})`

Explanation: Python functions are pass-by-object-reference. The value of `l` remains the same because integers are immutable. The list `m` and the dictionary `n` are mutable, so changes within the function reflect outside. `m` becomes `['x']` and `n` becomes `{'a': 'y'}`.
